```python
from PIL import Image
from paddleocr import PaddleOCR, draw_ocr, draw_structure_result
import csv
import cv2

# Initialize PaddleOCR with table structure model
ocr = PaddleOCR(
    use_angle_cls=True, 
    lang='en',  # Assuming the table model works with Chinese; change as needed
    use_space_char=True,  # Ensure space detection if needed
    table_model_dir='inference/ch_ppstructure_mobile_v2.0_SLANet_infer'
)

# Path to the image
img_path = 'table_image_cropped2.png'

# Perform OCR on the image
result = ocr.ocr(img_path, cls=True)

# Visualize recognized text on the image
image = Image.open(img_path).convert('RGB')
boxes = [line[0] for line in result[0]]
txts = [line[1][0] for line in result[0]]
scores = [line[1][1] for line in result[0]]

# Draw the OCR results on the image
im_show = draw_ocr(image, boxes, txts, scores, font_path='C:/Windows/Fonts/arial.ttf')
im_show = Image.fromarray(im_show)
im_show.save('result.jpg')

# Function to organize text into table rows and columns based on bounding box positions
def group_into_table(ocr_result):
    table_data = []
    current_row = []
    row_start_y = None

    # Sort the results by their vertical position
    sorted_results = sorted(ocr_result, key=lambda x: x[0][0][1])

    for line in sorted_results:
        text = line[1][0]
        top_left, top_right, bottom_right, bottom_left = line[0]
        top_y = top_left[1]
        bottom_y = bottom_left[1]
        x_positions = [top_left[0], top_right[0], bottom_left[0], bottom_right[0]]
        
        # Determine if it's a new row or column
        if row_start_y is None or abs(top_y - row_start_y) > 20:
            if current_row:
                table_data.append(current_row)
            current_row = []
            row_start_y = top_y

        current_row.append((x_positions, text))
    
    # Append the last row if it exists
    if current_row:
        table_data.append(current_row)

    # Convert row data to column-wise format
    table_transposed = []
    for row in table_data:
        max_columns = max(len(cell[0]) for cell in row)
        columns = [[] for _ in range(max_columns)]
        for cell in row:
            x_positions, text = cell
            column_index = min(range(len(x_positions)), key=lambda i: abs(x_positions[i] - (image.width / 2)))
            columns[column_index].append(text)
        table_transposed.append([' '.join(col) for col in columns])
    
    return table_transposed

# Organize OCR results into table rows and columns
table_data = group_into_table(result[0])

# Save the structured table to a CSV file
csv_file = 'table_output.csv'

with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)

    # Write each row into the CSV file
    for row in table_data:
        writer.writerow(row)

print(f"Table contents saved to {csv_file}")
```

```python
from PIL import Image
from paddleocr import PaddleOCR, draw_ocr
import csv
import numpy as np

# Initialize PaddleOCR with specific models
ocr = PaddleOCR(
    det_model_dir='inference/ch_PP-OCRv3_det_infer',  # Path to text detection model
    rec_model_dir='inference/ch_PP-OCRv3_rec_infer',  # Path to text recognition model
    table_model_dir='inference/ch_ppstructure_mobile_v2.0_SLANet_infer',  # Path to table structure model
    use_angle_cls=True,  # Use angle classifier for text rotation
    lang='en'  # Assuming the models are configured for English
)

# Path to the image
img_path = 'scr_1724963556.png'

# Perform OCR on the image
result = ocr.ocr(img_path, cls=True)

# Visualize recognized text on the image
image = Image.open(img_path).convert('RGB')
boxes = [line[0] for line in result[0]]
txts = [line[1][0] for line in result[0]]
scores = [line[1][1] for line in result[0]]

# Draw the OCR results on the image
im_show = draw_ocr(image, boxes, txts, scores, font_path='C:/Windows/Fonts/arial.ttf')
im_show = Image.fromarray(im_show)
im_show.save('result.jpg')

def group_into_table(ocr_result, image_width):
    table_data = []
    min_row_height = 20  # Threshold for determining a new row
    current_row = []
    
    # Sort the results by their vertical position
    sorted_results = sorted(ocr_result, key=lambda x: x[0][0][1])

    row_start_y = None

    for line in sorted_results:
        text = line[1][0]
        top_left, top_right, bottom_right, bottom_left = line[0]
        top_y = top_left[1]
        x_position = top_left[0]  # Use the x-coordinate of the top-left corner

        # Check if the current line starts a new row
        if row_start_y is None or abs(top_y - row_start_y) > min_row_height:
            if current_row:
                table_data.append(current_row)
            current_row = []
            row_start_y = top_y
        
        # Append the current text and x_position to the current row
        current_row.append((x_position, text))
    
    if current_row:
        table_data.append(current_row)
    
    # Determine column positions based on x-coordinates
    x_positions = [x for row in table_data for x, _ in row]
    x_positions = sorted(set(x_positions))  # Unique x-coordinates
    
    # Define column boundaries
    column_boundaries = np.linspace(min(x_positions), max(x_positions), num=len(x_positions) + 1)

    # Arrange data into columns
    table_transposed = []
    for row in table_data:
        columns = [''] * len(column_boundaries)
        for x_position, text in row:
            # Determine the column index based on x_position
            column_index = np.digitize(x_position, column_boundaries) - 1
            columns[column_index] += text + ' '
        table_transposed.append([col.strip() for col in columns])
    
    return table_transposed

# Organize OCR results into table rows and columns
table_data = group_into_table(result[0], image.width)

# Save the structured table to a CSV file
csv_file = 'table_output.csv'

with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)

    # Write each row into the CSV file
    for row in table_data:
        writer.writerow(row)

print(f"Table contents saved to {csv_file}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from PIL import Image
import os
import csv
from webdriver_manager.chrome import ChromeDriverManager
from paddleocr import PaddleOCR, draw_ocr
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up Chrome options
chrome_options = Options()
# Uncomment the next line if you want to see the browser
# chrome_options.add_argument('--headless')  # Uncomment this if you want to run in headless mode

# Initialize ChromeDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.set_window_size(1920, 1080)  # Set the window size

# Open the desired website
driver.get('https://dps.psx.com.pk/')  # Replace with your website URL

# Remove specific classes using JavaScript
class_to_remove_1 = 'header'  # Replace with the first class you want to remove
class_to_remove_2 = 'pageHeader'  # Replace with the second class you want to remove

# Execute JavaScript to remove both classes
driver.execute_script(f"""
    document.querySelectorAll('.{class_to_remove_1}').forEach(el => el.classList.remove('{class_to_remove_1}'));
    document.querySelectorAll('.{class_to_remove_2}').forEach(el => el.classList.remove('{class_to_remove_2}'));
""")

# Use explicit wait to ensure the table is loaded before taking the screenshot
wait = WebDriverWait(driver, 10)
table_element = wait.until(EC.presence_of_element_located((By.ID, 'DataTables_Table_0')))

# Take a screenshot of the table element
table_element.screenshot('scr1.png')

# Open the screenshot
image = Image.open('scr1.png')

# Save the screenshot
image.save('table_image_without_classes.png')

# Clean up the temporary file
os.remove('scr1.png')

# Define the path to the image
image_path = 'table_image_without_classes.png'
image = Image.open(image_path)
width, height = image.size

# Define the number of columns to crop from the right and the width of each column
columns_to_crop = 3.3
column_width = width // 10  # Adjust as needed

# Calculate the crop box
left = 0
top = 0
right = width - (columns_to_crop * column_width)
bottom = height

# Crop the image
cropped_image = image.crop((left, top, right, bottom))
cropped_image_path = 'table_image_cropped.png'
cropped_image.save(cropped_image_path)

# Initialize PaddleOCR with specific models
ocr = PaddleOCR(
    det_model_dir='inference/ch_PP-OCRv3_det_infer',  # Path to text detection model
    rec_model_dir='inference/ch_PP-OCRv3_rec_infer',  # Path to text recognition model
    table_model_dir='inference/ch_ppstructure_mobile_v2.0_SLANet_infer',  # Path to table structure model
    use_angle_cls=True,  # Use angle classifier for text rotation
    lang='en'  # Assuming the models are configured for English
)

# Perform OCR on the image
result = ocr.ocr(cropped_image_path, cls=True)

# Function to organize text into table rows and columns based on bounding box positions
def group_into_table(ocr_result, image_width):
    table_data = []
    min_row_height = 20  # Threshold for determining a new row
    current_row = []
    
    # Sort the results by their vertical position
    sorted_results = sorted(ocr_result, key=lambda x: x[0][0][1])

    row_start_y = None

    for line in sorted_results:
        text = line[1][0]
        top_left, top_right, bottom_right, bottom_left = line[0]
        top_y = top_left[1]
        x_position = top_left[0]

        if row_start_y is None or abs(top_y - row_start_y) > min_row_height:
            if current_row:
                table_data.append(current_row)
            current_row = []
            row_start_y = top_y
        
        current_row.append((x_position, text))
    
    if current_row:
        table_data.append(current_row)
    
    # Determine column positions based on x-coordinates
    column_positions = []
    for row in table_data:
        for x_position, _ in row:
            if not any(abs(x_position - pos) < image_width / 10 for pos in column_positions):
                column_positions.append(x_position)
    
    column_positions.sort()

    # Arrange data into columns
    table_transposed = []
    for row in table_data:
        columns = [''] * len(column_positions)
        for x_position, text in row:
            closest_column_index = min(range(len(column_positions)), key=lambda i: abs(x_position - column_positions[i]))
            columns[closest_column_index] += text + ' '
        table_transposed.append([col.strip() for col in columns])
    
    return table_transposed

# Organize OCR results into table rows and columns
table_data = group_into_table(result[0], image.width)

# Save the structured table to a CSV file
csv_file = 'table_output.csv'

with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    for row in table_data:
        writer.writerow(row)

print(f"Table contents saved to {csv_file}")

# The browser remains open; you can manually close it when done
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from PIL import Image
import os
import csv
import time
from webdriver_manager.chrome import ChromeDriverManager
from paddleocr import PaddleOCR
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up Chrome options
chrome_options = Options()
# Uncomment the next line if you want to see the browser
# chrome_options.add_argument('--headless')  # Uncomment this if you want to run in headless mode

# Initialize ChromeDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.set_window_size(1920, 1080)  # Set the window size

# Open the desired website
driver.get('https://dps.psx.com.pk/')  # Replace with your website URL

# Remove specific classes using JavaScript
class_to_remove_1 = 'header'
class_to_remove_2 = 'pageHeader'

driver.execute_script(f"""
    document.querySelectorAll('.{class_to_remove_1}').forEach(el => el.classList.remove('{class_to_remove_1}'));
    document.querySelectorAll('.{class_to_remove_2}').forEach(el => el.classList.remove('{class_to_remove_2}'));
""")

# Use explicit wait to ensure the table is loaded before taking screenshots
wait = WebDriverWait(driver, 10)
table_element = wait.until(EC.presence_of_element_located((By.ID, 'DataTables_Table_0')))

# Initialize PaddleOCR
ocr = PaddleOCR(
    det_model_dir='inference/ch_PP-OCRv3_det_infer',
    rec_model_dir='inference/ch_PP-OCRv3_rec_infer',
    table_model_dir='inference/ch_ppstructure_mobile_v2.0_SLANet_infer',
    use_angle_cls=True,
    lang='en',
    use_gpu=True  # Enable GPU
)

# CSV file to save OCR results
csv_file = 'table_output.csv'

# Open the CSV file for writing
with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)

    # Flag to check if headers have been extracted
    headers_extracted = False

    # Capture screenshots and process OCR
    capture_duration = 60  # Total duration to capture screenshots (in seconds)
    interval = 0.1  # Interval between screenshots (in seconds) - optimized for millisecond capture
    end_time = time.time() + capture_duration

    while time.time() < end_time:
        # Generate a timestamp for the file names
        timestamp = int(time.time() * 1000)  # Milliseconds

        # Take a screenshot of the table element
        screenshot_filename = f'scr_{timestamp}.png'
        table_element.screenshot(screenshot_filename)

        try:
            # Open the screenshot and process it
            with Image.open(screenshot_filename) as image:
                # Define the number of columns to crop from the right and the width of each column
                width, height = image.size
                columns_to_crop = 3.3
                column_width = width // 10  # Adjust as needed

                # Calculate the crop box
                left = 0
                top = 0
                right = width - (columns_to_crop * column_width)
                bottom = height

                # Crop the image
                cropped_image = image.crop((left, top, right, bottom))
                cropped_image_path = f'table_image_cropped_{timestamp}.png'
                cropped_image.save(cropped_image_path)

                try:
                    # Perform OCR on the cropped image
                    result = ocr.ocr(cropped_image_path, cls=True)

                    # Check if OCR result is None or empty
                    if result and result[0]:
                        # Function to organize text into table rows and columns based on bounding box positions
                        def group_into_table(ocr_result, image_width, ignore_first_row=False):
                            table_data = []
                            min_row_height = 25  # Threshold for determining a new row
                            current_row = []

                            # Sort the results by their vertical position
                            sorted_results = sorted(ocr_result, key=lambda x: x[0][0][1])

                            row_start_y = None
                            row_counter = 0

                            for line in sorted_results:
                                text = line[1][0]
                                top_left, top_right, bottom_right, bottom_left = line[0]
                                top_y = top_left[1]
                                x_position = top_left[0]

                                if row_start_y is None or abs(top_y - row_start_y) > min_row_height:
                                    if current_row:
                                        if not ignore_first_row or row_counter > 0:
                                            table_data.append(current_row)
                                        current_row = []
                                    row_start_y = top_y
                                    row_counter += 1
                                
                                current_row.append((x_position, text))
                            
                            if current_row:
                                if not ignore_first_row or row_counter > 0:
                                    table_data.append(current_row)
                            
                            # Determine column positions based on x-coordinates
                            column_positions = []
                            for row in table_data:
                                for x_position, _ in row:
                                    if not any(abs(x_position - pos) < image_width / 10 for pos in column_positions):
                                        column_positions.append(x_position)
                            
                            column_positions.sort()

                            # Arrange data into columns
                            table_transposed = []
                            for row in table_data:
                                columns = [''] * len(column_positions)
                                for x_position, text in row:
                                    closest_column_index = min(range(len(column_positions)), key=lambda i: abs(x_position - column_positions[i]))
                                    columns[closest_column_index] += text + ' '
                                table_transposed.append([col.strip() for col in columns])
                            
                            return table_transposed

                        # Organize OCR results into table rows and columns
                        ignore_first_row = headers_extracted
                        table_data = group_into_table(result[0], image.width, ignore_first_row)

                        # If headers haven't been extracted yet, extract them and write to CSV
                        if not headers_extracted:
                            headers = table_data[0] if table_data else []
                            writer.writerow(headers)
                            headers_extracted = True

                        # Append the table data (excluding headers) to the CSV file
                        for row in table_data[1:]:  # Skip the first row if it's the header
                            writer.writerow(row)

                    else:
                        print(f"No OCR results for {cropped_image_path}. Skipping this image.")
                
                finally:
                    # Clean up temporary cropped image file
                    os.remove(cropped_image_path)

        finally:
            # Clean up temporary screenshot file
            os.remove(screenshot_filename)

        # Wait for the next interval
        time.sleep(interval)

# The browser remains open; you can manually close it when done
print(f"Table contents saved to {csv_file}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from PIL import Image
import io
import csv
import time
from webdriver_manager.chrome import ChromeDriverManager
from paddleocr import PaddleOCR
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from concurrent.futures import ThreadPoolExecutor
from selenium.common.exceptions import NoSuchWindowException, WebDriverException

# Set up Chrome options
chrome_options = Options()
# Uncomment the next line if you want to see the browser
# chrome_options.add_argument('--headless')  # Uncomment this if you want to run in headless mode

# Initialize ChromeDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.set_window_size(1920, 1080)  # Set the window size

# Open the desired website
driver.get('https://dps.psx.com.pk/')  # Replace with your website URL

# Remove specific classes using JavaScript (run once)
class_to_remove_1 = 'header'
class_to_remove_2 = 'pageHeader'

driver.execute_script(f"""
    document.querySelectorAll('.{class_to_remove_1}').forEach(el => el.classList.remove('{class_to_remove_1}'));
    document.querySelectorAll('.{class_to_remove_2}').forEach(el => el.classList.remove('{class_to_remove_2}'));
""")

# Use explicit wait to ensure the table is loaded before taking screenshots
wait = WebDriverWait(driver, 10)
table_element = wait.until(EC.presence_of_element_located((By.ID, 'DataTables_Table_0')))

# Initialize PaddleOCR
ocr = PaddleOCR(
    det_model_dir='inference/ch_PP-OCRv3_det_infer',
    rec_model_dir='inference/ch_PP-OCRv3_rec_infer',
    table_model_dir='inference/ch_ppstructure_mobile_v2.0_SLANet_infer',
    use_angle_cls=True,
    lang='en'
)

# CSV file to save OCR results
csv_file = 'table_output.csv'

# Open the CSV file for writing
with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)

    # Flag to check if headers have been extracted
    headers_extracted = False

    # Capture screenshots and process OCR
    capture_duration = 60  # Total duration to capture screenshots (in seconds)
    interval = 0.1  # Interval between screenshots (in seconds) - optimized for millisecond capture
    end_time = time.time() + capture_duration

    # ThreadPoolExecutor to process OCR in parallel
    with ThreadPoolExecutor(max_workers=2) as executor:
        futures = []

        while time.time() < end_time:
            start_time = time.perf_counter()

            try:
                # Capture screenshot directly to memory
                screenshot_binary = table_element.screenshot_as_png

                def process_image(screenshot_binary):
                    try:
                        # Open screenshot from memory
                        image = Image.open(io.BytesIO(screenshot_binary))

                        # Define the number of columns to crop from the right and the width of each column
                        width, height = image.size
                        columns_to_crop = 3.3
                        column_width = width // 10  # Adjust as needed

                        # Calculate the crop box
                        left = 0
                        top = 0
                        right = width - int(columns_to_crop * column_width)
                        bottom = height

                        # Crop the image
                        cropped_image = image.crop((left, top, right, bottom))

                        # Perform OCR on the cropped image
                        result = ocr.ocr(cropped_image, cls=True)

                        if result and result[0]:
                            # Function to organize text into table rows and columns based on bounding box positions
                            def group_into_table(ocr_result, image_width, ignore_first_row=False):
                                table_data = []
                                min_row_height = 25  # Threshold for determining a new row
                                current_row = []

                                # Sort the results by their vertical position
                                sorted_results = sorted(ocr_result, key=lambda x: x[0][0][1])

                                row_start_y = None
                                row_counter = 0

                                for line in sorted_results:
                                    text = line[1][0]
                                    top_left, top_right, bottom_right, bottom_left = line[0]
                                    top_y = top_left[1]
                                    x_position = top_left[0]

                                    if row_start_y is None or abs(top_y - row_start_y) > min_row_height:
                                        if current_row:
                                            if not ignore_first_row or row_counter > 0:
                                                table_data.append(current_row)
                                            current_row = []
                                        row_start_y = top_y
                                        row_counter += 1

                                    current_row.append((x_position, text))

                                if current_row:
                                    if not ignore_first_row or row_counter > 0:
                                        table_data.append(current_row)

                                # Determine column positions based on x-coordinates
                                column_positions = []
                                for row in table_data:
                                    for x_position, _ in row:
                                        if not any(abs(x_position - pos) < image_width / 10 for pos in column_positions):
                                            column_positions.append(x_position)

                                column_positions.sort()

                                # Arrange data into columns
                                table_transposed = []
                                for row in table_data:
                                    columns = [''] * len(column_positions)
                                    for x_position, text in row:
                                        closest_column_index = min(range(len(column_positions)),
                                                                   key=lambda i: abs(x_position - column_positions[i]))
                                        columns[closest_column_index] += text + ' '
                                    table_transposed.append([col.strip() for col in columns])

                                return table_transposed

                            # Organize OCR results into table rows and columns
                            ignore_first_row = headers_extracted
                            table_data = group_into_table(result[0], cropped_image.width, ignore_first_row)

                            # If headers haven't been extracted yet, extract them and write to CSV
                            if not headers_extracted:
                                headers = table_data[0] if table_data else []
                                writer.writerow(headers)
                                headers_extracted = True

                            # Append the table data (excluding headers) to the CSV file
                            for row in table_data[1:]:  # Skip the first row if it's the header
                                writer.writerow(row)

                        else:
                            print(f"No OCR results for the image. Skipping this image.")

                    except Exception as e:
                        print(f"Error processing image: {e}")

                # Process the image in a separate thread
                futures.append(executor.submit(process_image, screenshot_binary))

            except (NoSuchWindowException, WebDriverException) as e:
                print(f"Browser window closed or unavailable: {e}")
                break

            # Wait for precise interval
            elapsed_time = time.perf_counter() - start_time
            if elapsed_time < interval:
                time.sleep(interval - elapsed_time)

        # Wait for all OCR tasks to finish
        for future in futures:
            future.result()

# The browser remains open; you can manually close it when done
print(f"Table contents saved to {csv_file}")
```
